```python
import streamlit as st
import pandas as pd
import plotly.express as px

# Title and User Profile
st.set_page_config(page_title="Fresh Hearth Dashboard")
st.title("ü•ó Fresh Hearth - Personalized Nutrition Dashboard")
st.sidebar.header("üë§ Your Profile")
st.sidebar.text("Name: Ethan")
st.sidebar.text("Goal: Improve Gut Health")
st.sidebar.text("Level: Wellness Warrior")

# Points and Progress
st.subheader("üåü Your Health Points")
st.metric("Total Points", "üèÖ 1,230")
st.progress(0.75)  # 75% towards next reward tier

# Rewards Section
st.subheader("üéÅ Rewards Center")
st.markdown("- ‚úÖ Completed 7-day meal tracking")
st.markdown("- üîì Unlocked: Hydration Hero Badge")
st.markdown("- üéØ 3 days left to complete 'Meatless Monday Challenge'")

# Personalized Feedback Quiz
st.subheader("üß† Weekly Feedback Quiz")
if st.button("Take Quiz"):
    st.info("This week's quiz will ask about your energy levels and food satisfaction.")

# Meal Tracking Chart
st.subheader("üìä Your Weekly Calorie Intake")
data = pd.DataFrame({
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Calories": [1800, 1900, 2000, 1750, 1850, 2200, 2100]
})
fig = px.line(data, x="Day", y="Calories", markers=True, title="Calorie Intake Over the Week")
st.plotly_chart(fig)

# Personalized Insights
st.subheader("üí° Smart Insights")
st.info("üîé You tend to consume higher calories on weekends. Consider lighter dinner options.")
st.info("ü•ó You skipped breakfast 3 times this week. Want to set a morning reminder?")

# Footer
st.markdown("---")
st.caption("Fresh Hearth - Empowering Healthy Living Through Data")
```
